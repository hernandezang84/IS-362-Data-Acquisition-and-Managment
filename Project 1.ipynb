```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Reads the information from .CSV file into pandas
df = pd.read_csv('C:/Users/Jessie/Documents/IS 362/project1.csv')
```

```python
# Prints the first few rows of the data
print(df.head())
```

```python
# Calculates the delay rate for each airline
df['Alaska_Delay_Rate'] = df['Alaska_Delay'] / (df['Alaska_Delay'] + df['Alaska_On_Time'])
df['AM_WEST_Delay_Rate'] = df['AM WEST_Delay'] / (df['AM WEST_Delay'] + df['AM WEST_On_Time'])

print(df)
```

```python
print(df[['CITY', 'Alaska_Delay_Rate', 'AM_WEST_Delay_Rate']])
```

```python
# Plots the delay rate for each airline
plt.figure(figsize=(10,6))
plt.bar(df['CITY'], df['Alaska_Delay_Rate'], width=0.4, label='Alaska', color='b', align='center')
plt.bar(df['CITY'], df['AM_WEST_Delay_Rate'], width=0.4, label='AM WEST', color='r', align='edge')

plt.xlabel('CITIES')
plt.ylabel('Delay Rate')
plt.title('Comparison of Arrival Delays between Alaska and AM West Airlines')
plt.legend()
plt.show()
```

# Comparison

According to the graph, we can see that AM WEST has more arrival delays in all cities compared to Alaska.  Given this information, it would be better to fly Alaska airlines rather than AM WEST airlines.
