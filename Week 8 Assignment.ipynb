```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
column_names = ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'model year', 'origin', 'car name']
```

```python
df = pd.read_csv("C:/Users/Jessie/Documents/IS 362/auto-mpg/auto-mpg.data", delim_whitespace=True, names=column_names)
```

```python
df['horsepower'] = pd.to_numeric(df['horsepower'], errors='coerce')
```

```python
origin_map = {1: 'USA', 2: 'Asia', 3: 'Europe'}
df['origin'] = df['origin'].map(origin_map)
```

```python
sns.set(style="whitegrid")
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='cylinders', data=df)
plt.title('Distribution of Cylinders in the Dataset')
plt.xlabel('Number of Cylinders')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='horsepower', y='weight', data=df)
plt.title('Relationship Between Horsepower and Weight')
plt.xlabel('Horsepower')
plt.ylabel('Weight')
plt.show()
```

Q: How does the fuel efficiency (mpg) vary with the number of cylinders and the vehicle's weight?

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='cylinders', y='weight', size='mpg', hue='mpg', data=df, sizes=(20, 200), alpha=0.6)
plt.title('Fuel Efficiency Variation with Cylinders and Weight')
plt.xlabel('Number of Cylinders')
plt.ylabel('Weight')
plt.legend(title='MPG', bbox_to_anchor=(1.05, 1), loc=2)
plt.show()
```

A: The scatterplot illustrates how fuel efficiency (mpg) varies with the number of cylinders and vehicle weight.  In the plot, the size and color intensity represent the MPG values, providing insights into how these factors interplay to affect fuel efficiency.
